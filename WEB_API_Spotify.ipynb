```python
artist_name = response['albums']['items'][0]['artists'][0]['name']
artist_name
```

```python
no_of_tracks = response['albums']['items'][0]['total_tracks']
no_of_tracks
```

```python
#Converting data into a list

new_release_data = []
new_release_data.append(response)
```

```python
len(response['albums']['items'])
```

```python
# For Loop

artist_name = []
album_name = []
album_release_date = []
no_of_tracks = []

for data in new_release_data:
    for i in range(0,len(response['albums']['items'])):
        artist_name.append(data['albums']['items'][i]['artists'][0]['name'])
        album_name.append(data['albums']['items'][i]['name'])
        album_release_date.append(data['albums']['items'][i]['release_date'])
        no_of_tracks.append(data['albums']['items'][i]['total_tracks'])
    
print ("The artist name: " + str(artist_name))
print ("The album name: " + str(album_name))
print ("The album release date: " + str(album_release_date))
print ("The no. of tracks: " + str(no_of_tracks))
```

```python
# Create a dictionary containing our newly extracted information

final_data = {"Artist_Name": artist_name, 
                "Album name": album_name,
                "Album Release Date": album_release_date,
                "No. of tracks": no_of_tracks}
```

```python
# Convert our dictionary into a Pandas Data Frame
import pandas as pd

final_data = pd.DataFrame(final_data).reset_index()
final_data
```
